```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import iris data set
iris = pd.read_csv('Iris.csv')
print(iris.head())
print(iris.duplicated().sum())
print(iris.isnull().sum())
print(iris.describe())
```

```python
iris_df = iris.drop(columns=['Id', 'Species'])

fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(12, 12))
axes = axes.flatten()

bins_values = [5, 10, 15, 20]

for i, feature in enumerate(iris_df.columns):
    for j, bins in enumerate(bins_values):
        sns.histplot(iris_df[feature],ax=axes[i * 4 + j], bins=bins)
        axes[i * 4 + j].set_title(f"{feature} (Bins: {bins})")

plt.tight_layout()
plt.show()
```
